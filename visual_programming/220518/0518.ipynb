```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')


# make data
x = [1, 2, 3, 4]
colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(x)))

# plot
fig, ax = plt.subplots()
ax.pie(x, colors=colors, radius=3, center=(4, 4),
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
x = [x/10 for x in range(20)]     # 0 20  
y = [(x/ 10)**2 for x in range(20)]  # 0 20  x  x  
z = [(x/10)**3 for x in range(20)]  # 0 20  x  x   
i = [2**(x/10) for x in range(20)]

plt.figure(figsize=(10, 5))
plt.plot(x, x, color='r', ls='-', lw=2, label='linear')    #    
plt.plot(x, y, color='g', ls='--', lw=3, label='quadratic') 
plt.plot(x, z, color='b', ls='-.', lw=4, label='cubic') 
plt.plot(x, i, color='k', ls=':', lw=5, label='power') 

plt.xlabel('x label')      # x  
plt.ylabel('y label')      # y  
plt.title("My Plot") 
plt.legend() 
plt.show() 
```

```python
x, y1, y2 = [], [], []

for i in range(0, 50, 2):
    x.append(i/10)
    y1.append((i/10)**2)
    y2.append((i/10)**3)

plt.figure(figsize=(10, 5))
plt.plot(x, x, 'ro', x, y1, 'bs', x, y2, 'g^')
plt.plot(x, x, label='First Line')
plt.plot(x, y1, label='Second Line')
plt.plot(x, y2, label='Third Line')

plt.xlabel('Plot Number')
plt.ylabel('Important var')
plt.title('Interesting Graph\nCheck it out')
plt.legend()
plt.show()
```

```python
x = np.arange(0.0, 2*np.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(2, 1, 1)
plt.plot(x, y1, 'r-')
plt.title('1st Graph')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x, y2, 'b-.')
plt.title('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

```python
x1 = np.linspace(0, 5)
x2 = np.linspace(0, 2)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'ro-')
plt.title('1st Graph')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'b-.')
plt.title('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

```python
x1 = np.linspace(0, 5)
x2 = np.linspace(0, 2)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(1, 2, 1)
plt.plot(x1, y1, 'ro-')
plt.title('1st Graph')
plt.ylabel('Damped oscillation')

plt.subplot(1, 2, 2)
plt.plot(x2, y2, 'b-.')
plt.title('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

```python
x = np.arange(0, 2*np.pi, 0.1)
sin_y = np.sin(x)
cos_y = np.cos(x)

fig, ax = plt.subplots(2, 2, figsize=(5, 5))

ax[0, 0].plot(x, sin_y, 'r') #row=0, col=0
ax[1, 0].plot(x, 1-sin_y, 'b') #row=1, col=0
ax[0, 1].plot(x, cos_y, 'g') #row=0, col=1
ax[1, 1].plot(x, 1-cos_y, 'k') #row=1, col=1
plt.show()
```

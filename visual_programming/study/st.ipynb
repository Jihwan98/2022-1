```python
import csv

f = open('weather.csv', 'r', encoding='cp949')

data = csv.reader(f)
header = next(data)
print(header)

max_wind = 0
for d in data:
    if d[2] != '':
        max_wind = max(max_wind, float(d[2]))
        max_day = d[0]
print("최대 풍속 :", max_wind, ", 날짜 :", max_day)

f.close()
```

```python
f = open('weather.csv', 'r', encoding='cp949')

data = csv.reader(f)
header = next(data)

months = [i for i in range(1, 13)]
winds = [0 for _ in range(12)]
days_count = [0 for _ in range(12)]

for d in data:
    if d[3] != '':
        month = int(d[0].split('-')[1])
        winds[month-1] += float(d[3])
        days_count[month-1] += 1

for idx, z in enumerate(zip(winds, days_count)):
    w, d = z
    winds[idx] = w/d

plt.plot(winds)
plt.xticks(range(len(winds)), months)
plt.show()

f.close()
```

```python
from sklearn import linear_model

regr = linear_model.LinearRegression()

X = [[164, 1], [167, 1], [165, 0], [170, 0], [179, 0], [163, 1], [159, 0], [166, 1]]    # 입력데이터를 2차원으로 만들어야 함 
y = [43, 48, 47, 66, 67, 50, 52, 44]     # y 값은 1차원 데이터

regr.fit(X, y)
print("intercept :", regr.intercept_)
print("coef :", regr.coef_)
print("score :", regr.score(X, y))
print("regr([166, 0]) :", regr.predict([[166, 0]]), "regr([166, 1]) :", regr.predict([[166, 1]]))
```

```python
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(X, y, test_size = 0.2)
train_x, test_x, train_y, test_y
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split

iris = datasets.load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2)

from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
scores = metrics.accuracy_score(y_test, y_pred)
print(scores)
```

```python
from sklearn.cluster import KMeans

iris = datasets.load_iris()
X2 = iris.data[:, 2:4]
y = iris.target

kmeans = KMeans(n_clusters=3, random_state=21)
kmeans.fit(X2)
centers = kmeans.cluster_centers_
print(centers)

plt.scatter(X2[:, 0], X2[:, 1], c=y)
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn import datasets
import matplotlib.pyplot as plt

x, y = datasets.make_blobs(n_samples=300, centers=3, cluster_std=0.6, random_state=0)
kmeans = KMeans(n_clusters=3)
kmeans.fit(x)
y_kmeans = kmeans.predict(x)
centers = kmeans.cluster_centers_
plt.scatter(x[:,0], x[:,1], c=y, s=50, cmap='rainbow')
plt.scatter(centers[:,0], centers[:,1], c='black', s=200, alpha=0.5)
plt.show()
```

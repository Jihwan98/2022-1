```python
import math 
import matplotlib.pyplot as plt 
 
x = []
sin = [] 
 
for angle in range(360): 
    x.append(angle) 
    sin.append(math.sin(math.radians(angle))) 
 
plt.plot(x, sin) 
plt.title("SINE WAVE") 
plt.show()
```

```python
import math 
import matplotlib.pyplot as plt 
 
x = [] 
sin = [] 
cos = []
 
for angle in range(360): 
    x.append(angle) 
    sin.append(math.sin(math.radians(angle))) 
    cos.append(math.cos(math.radians(angle)))
 
plt.plot(x, sin, "b:", label='sin')
plt.plot(x, cos, "r-.", label='cos')
plt.legend()
plt.title("SINE WAVE and COSINE WAVE") 
plt.show()
```

```python
import math 
import matplotlib.pyplot as plt 
 
x = [] 
sin = [] 
cos = []
 
for angle in range(0, 360, 10):
    x.append(angle) 
    sin.append(math.sin(math.radians(angle))) 
    cos.append(math.cos(math.radians(angle)))

plt.plot(x, sin, "bx-.", label='sin')
plt.plot(x, cos, "r*:", label='cos')
plt.legend()
plt.title("SINE WAVE and COSINE WAVE") 
plt.show()
```
